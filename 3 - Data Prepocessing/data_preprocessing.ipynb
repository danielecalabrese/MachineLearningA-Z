## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing dataset

```python
dataset = pd.read_csv('Data.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
print(X)
```

```python
y = dataset.iloc[:, -1].values
print(y)
```

## Cleaning data: Taking care of missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3]) # aggiungendo 1:3 escludi la prima colonna (quella con i Paesi)
X[:, 1:3] = imputer.transform(X[:, 1:3])
print(X)
```

## Encoding categorical variables

### Encoding Indipendent variable

```python
from sklearn.compose import ColumnTransformer # ColumnTransformer è una classe che permette di applicare trasformazioni specifiche a colonne specifiche del dataset. In questo caso, stiamo usando ColumnTransformer per applicare OneHotEncoder alla prima colonna (colonna 0) del dataset X, mentre le altre colonne rimangono invariate (remainder='passthrough'). Si usa per la variabile X, non per la variabile Y
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
print(X)
```

### Encoding Dependent variable

```python
from sklearn.preprocessing import LabelEncoder # Label Endoder trasforma le stringhe in numeri, ma non crea nuove colonne come OneHotEncoder. Si usa per la varialbile dipendente, non per quella indipendente
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```
