## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing dataset

```python
dataset = pd.read_csv('Salary_Data.csv') # il csv ha due colonne: Anni di esperienza e Stipendio
dataset.head()
```

```python
X = dataset.iloc[:, 0].values
y = dataset.iloc[:, 1].values
```

```python
print(X)
```

```python
print(y)
```

## Splitting dataset into Test set and Train set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
X_train = X_train.reshape(-1, 1) # il modello di regressione lineare di sklearn si aspetta un array 2D, quindi dobbiamo trasformare X_train in una matrice con una sola colonna. Puoi controllare le dimensioni con print(X_train.shape). Questa funziona non modifica X_train, ma restituisce una nuova matrice con le dimensioni corrette.
X_test = X_test.reshape(-1, 1) # lo stesso per X_test
print(X_train) 
```

## Training the Simple Regression model on train set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

## Visualize the Training set results

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Stipendio vs Anni di esperienza (Training set)')
plt.xlabel('Anni di esperienza')
plt.ylabel('Stipendio')
plt.show()
```
